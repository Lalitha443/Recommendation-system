```python
#  3. Create User-Item Matrix
user_df = df.pivot(index='anime_id',
                                 columns='name',
                                 values='rating')
user_df
```

```python
#Impute those NaNs with 0 values
df["genre"].fillna(0, inplace=True)
df["type"].fillna(0, inplace=True)
df["rating"].fillna(0, inplace=True)
```

```python
#  4. Compute User Similarity

# from scipy.spatial.distance import cosine correlation
# Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
user_sim = 1 - pairwise_distances(user_df.values,metric='cosine')

#pairwise_distances(user_df.values,metric='cosine')


user_sim
```

```python
# Fill NaN values in user_df with 0
user_df.fillna(0, inplace=True)
```

```python
import numpy as np
np.fill_diagonal(user_sim, 0)


user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids
user_sim_df.index   = df.anime_id.unique()
user_sim_df.columns = df.anime_id.unique()

user_sim_df

```

```python
# 6. Identify Most Similar Users

#Most Similar Users
user_sim_df.max()

user_sim_df.idxmax(axis=1)[0:10]

```
